# Operadores aritméticos y relacionales
Los operadoes son las herramientas básicas para hacer las operaciones. Los aritméticos(+, -, *, /) se pueden usar para sumar, restar, multiplicar, promedior, etc. Pero no son los unicos tambien estan los operadores racionales (<, >, =), los cuales nos permiten sabes si los resultadso son iguales o un número es mayor que otro.

```r
a = 10
b = 3

print(a + b)
print(a ** 2)       # Potencia
print(a %% b)       # Módulo  Se utiliza para determinar si un número es par o impar
print(a > b)
```

# **Vectores**

En esatdistica rara vez trabajamos con un solo número. Normalmente tenemos muchos números: los resultados de 100 analisís de glucosa, las alturas de 50 personas, etc. Un vector es precisamente eso: una colección ordenada de números.
Con los vectores podemos calcular el promedio, el máximo, el mínimo o la suma de todos los valores de una sola vez.

```r
edades <- c(23, 25, 19, 34, 28)
glucosas <- c(89, 110, 145, 98, 180)

edades
mean(edades)
median(edades)
sd(edades)
range(edades)
```

# **Comparaciones lógicas**

```r
glucosas > 126
sum(glucosas > 126)
mean(glucosas > 126)

# Interpretación:
# sum() cuenta cuántos cumplen la condición.
# mean() calcula la proporción TRUE si TRUE=1 y FALSE=0.
```

# **Data frames: tabla de pacientes**

En la vida real, los datos casi nunca vienen como vectores aislados. Normalmente tenemos una tabla donde cada fila es una observación (por ejemplo, un estudiante) y cada columna es una variable (nombre, edad, nota). Esto es exactamente un Data Frame. Aprender a manejar data frames es fundamental porque la mayoría de los problemas de estadística involucran datos organizados de esta forma.

*   Elemento de la lista
*   Elemento de la lista



```r
datos <- data.frame(
    Nombre = c("Ana", "Luis", "María", "José"),
    Edad = c(22, 25, 20, 28),
    Nota = c(9.2, 8.5, 7.8, 9.0)
)
print(datos)


#Ejemplo 2
edades <- c(23, 25, 19, 34, 28)
glucosas <- c(89, 110, 145, 98, 180)

pacientes <- data.frame(
  id = 1:5,
  edad = edades,
  sexo = c("F", "M", "F", "M", "F"),
  glucosa = glucosas,
  diabetes = glucosas >= 126
)

pacientes
summary(pacientes)
```

#**Factores**

```r
pacientes$sexo <- factor(pacientes$sexo, levels = c("F", "M"))
pacientes$diabetes <- factor(
  pacientes$diabetes,
  levels = c(FALSE, TRUE),
  labels = c("No", "Sí")
)


pacientes
str(pacientes)
summary(pacientes)
```

#**Seleccionar columnas y filas con base R**

```r
pacientes$edad
pacientes[1, ]
pacientes[, "glucosa"]
pacientes[pacientes$glucosa >= 126, ]
```
