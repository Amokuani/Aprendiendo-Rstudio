```r
#cargar paquetes
install.packages(c("readxl","corrplot","RColorBrewer"))

library(readxl)
library(corrplot)
library(RColorBrewer)
```

```r
#Cargar datos
datos <- read_excel("datos mapa calor.xlsx")

head(datos)

datos_num <- datos[, -1]

str(datos)
```

```r
#Calcular correlaciones
correlaciones <- cor(datos_num)

round(correlaciones,2)
```

```r
#Mapa de calor
corrplot(
  correlaciones,
  method = "color",
  addCoef.col = "black",
  number.cex = 0.8,
  tl.col = "black",
  tl.srt = 45,
  col = colorRampPalette(c("blue","white","red"))(200)
)

# Interpertación
#Azul = correlación negativa.
#Rojo = correlación positiva.
#Blanco = correlación cercana a cero.
#Los números corresponden a rrr de Pearson.
```

```r
#Versión más elegante
corrplot(
  correlaciones,
  method = "square",
  type = "upper",
  addCoef.col = "black",
  tl.col = "black",
  tl.cex = 0.9,
  number.cex = 0.7,
  diag = FALSE,
  col = colorRampPalette(c("#2166AC","white","#B2182B"))(100)
)
```

```r
# Con ggplot 2
install.packages("reshape2")
library(reshape2)
library(ggplot2)

cor_mat <- cor(datos_num)

cor_df <- melt(cor_mat)

ggplot(cor_df,
       aes(Var1, Var2, fill = value)) +

  geom_tile(color = "white") +

  geom_text(aes(label = round(value,2)),
            size = 4,
            color = "black") +

  scale_fill_gradient2(
    low = "blue",
    mid = "white",
    high = "red",
    midpoint = 0,
    limits = c(-1,1)
  ) +

  theme_minimal() +

  theme(
    axis.text.x = element_text(
      angle = 45,
      hjust = 1
    )
  ) +

  labs(
    title = "Mapa de calor de correlaciones",
    fill = "r"
  )
```

```r
#Mapa
install.packages(
  c("sf","dplyr","ggplot2",
    "rnaturalearth","rnaturalearthdata")
)


library(sf)
library(ggplot2)
library(dplyr)
```

```r
#Cargar Shp
mexico <- st_read("mexico.shp")
```

```r
#Unir datos tabla
datos_mapa <- data.frame(
  Estado = datos[[1]],
  alacranismo = datos$`Alacranismo H`
)
```

```r
#Unir datos shp
mexico2 <- left_join(
  mexico,
  datos_mapa,
  by = c("ESTADO" = "Estado")
)
```

```r
# mapa
ggplot(mexico2) +

  geom_sf(
    aes(fill = alacranismo),
    color = "white"
  ) +

  scale_fill_gradient(
    low = "#2C7BB6",
    high = "#D7191C",
    name = "Casos"
  ) +

  theme_minimal() +

  labs(
    title = "Alacranismo masculino por estado",
    subtitle = "México"
  )
```
