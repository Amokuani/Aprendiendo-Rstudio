#**Gráficos clínicos con ggplot2**


Importancia de las gráficas en el análisis de datos
En el análisis de datos, las gráficas son una herramienta fundamental porque permiten transformar números en información visual fácil de interpretar. Aunque los estadísticos y tablas son importantes, muchas veces no son suficientes para entender rápidamente lo que está ocurriendo en los datos. Las visualizaciones nos ayudan a identificar patrones, tendencias, diferencias entre grupos y posibles valores atípicos de una manera mucho más intuitiva.
En el contexto clínico, esto es especialmente relevante, ya que facilita la toma de decisiones. Por ejemplo, una gráfica puede mostrar de forma clara si existen diferencias entre grupos de tratamiento, cómo se distribuyen variables como la glucosa o si hay relación entre factores como el IMC y la presión arterial. De esta manera, las gráficas no solo ayudan a analizar, sino también a comunicar resultados de forma efectiva a otros profesionales de la salud.
En R, el paquete ggplot2 permite crear gráficos de alta calidad de forma estructurada y flexible. A través de diferentes tipos de visualizaciones —como gráficos de barras, histogramas, boxplots y gráficos de dispersión— podemos explorar distintos aspectos de los datos. Además, herramientas como las facetas permiten comparar múltiples variables al mismo tiempo.
Otro punto clave es que las gráficas forman parte del flujo de trabajo reproducible: podemos generarlas automáticamente y guardarlas como archivos (por ejemplo, en formato .png), lo que facilita su uso en presentaciones, artículos o reportes clínicos.

```r
#Cargar librerias

install.packages("tidyverse")
install.packages("janitor")
install.packages("skimr")

library(tidyverse)
library(janitor)

#Cargar datos
datos <- read_csv("data/datos_medicos.csv") %>%
  clean_names() %>%
  mutate(
    sexo = factor(sexo),
    grupo = factor(grupo),
    diabetes = factor(diabetes, levels = c(0, 1), labels = c("No", "Sí")),
    hipertension = factor(hipertension, levels = c(0, 1), labels = c("No", "Sí")),
    evento_complicacion = factor(evento_complicacion, levels = c(0, 1), labels = c("No", "Sí"))
  )

dir.create("figuras", showWarnings = FALSE)
```

#**1. Gráfico de barras: frecuencia por grupo**

El gráfico de barras es una de las herramientas más básicas y útiles en el análisis de datos. Se utiliza para representar la frecuencia o cantidad de observaciones en distintas categorías. En este tipo de gráfico, cada barra representa una categoría, y su altura indica cuántos casos pertenecen a dicha categoría.
Este tipo de visualización es especialmente útil cuando trabajamos con variables categóricas, como grupos de tratamiento, sexo, presencia o ausencia de una enfermedad, entre otros. Por ejemplo, en un contexto clínico, un gráfico de barras permite ver rápidamente cuántos pacientes hay en cada grupo o cuántos presentan una determinada condición.
El gráfico de barras sirve principalmente para:

Resumir datos categóricos de forma visual
Comparar frecuencias entre diferentes grupos
Identificar desequilibrios en la distribución de los datos

Se recomienda utilizar gráficos de barras cuando:

La variable es cualitativa o categórica
Se desea comparar cantidades entre grupos
Se busca una visualización simple, clara y fácil de interpretar

En resumen, el gráfico de barras es una herramienta fundamental para el análisis descriptivo inicial, ya que permite obtener una visión general rápida de cómo están distribuidos los datos en categorías.

```r
g_barras <- ggplot(datos, aes(x = grupo)) +
  geom_bar() +
  labs(
    title = "Número de pacientes por grupo",
    x = "Grupo",
    y = "Frecuencia"
  ) +
  theme_minimal()

g_barras

ggsave("figuras/01_barras_grupo.png", g_barras, width = 7, height = 5, dpi = 300)
```

#**¿Cómo cambiar los colores de las barras en ggplot2?**

 1.  **Añade el argumento `fill`:** Dentro de `aes()` en `ggplot()`, incluye `fill = tu_variable_categorica`. Esto le dice a ggplot2 que use los valores de esa variable para rellenar las barras con colores diferentes.
    Ejemplo: `ggplot(datos, aes(x = grupo, fill = grupo))`

 2.  **Usa `scale_fill_manual()`:** Esta función te permite especificar exactamente qué color quieres para cada categoría.

     **Opción A: Usando NOMBRES de colores:**
     R reconoce muchos nombres de colores en inglés (por ejemplo, "red", "blue", "green", "purple", "steelblue", "darkgreen", "orange", etc.).
     Puedes asignarlos directamente a tus grupos:
     `scale_fill_manual(values = c("NombreDelGrupo1" = "nombre_color1", "NombreDelGrupo2" = "nombre_color2"))`
     ¡Intenta buscar una lista de nombres de colores en R para más opciones!

     >https://r-charts.com/es/colores/

```r
g_barras_colores_nombre <- ggplot(datos, aes(x = grupo, fill = grupo)) + # Añadimos 'fill = grupo' para que las barras se coloreen según cada grupo
  geom_bar() +
  labs(
    title = "Número de pacientes por grupo (con nombres de colores)",
    x = "Grupo",
    y = "Frecuencia"
  ) +
  theme_minimal() +
  # Usa scale_fill_manual para asignar colores. Aquí usamos nombres de colores.
  # Asegúrate de que los nombres de los grupos coincidan con los de tus datos.
  scale_fill_manual(values = c("Control" = "steelblue", "Tratamiento_A" = "darkgreen", "Tratamiento_B" = "orange")) # Ejemplo: cambiar "Control", "Tratamiento A", etc. por los nombres reales de tus grupos

g_barras_colores_nombre # Muestra el gráfico con colores por nombre

ggsave("figuras/02_barras_grupo_colores.png", g_barras_colores_nombre, width = 7, height = 5, dpi = 300) # Guarda el gráfico
```

#**Gráfico de barras apiladas: proporción de diabetes por grupo**

El gráfico de barras apiladas es una extensión del gráfico de barras tradicional que permite no solo visualizar frecuencias, sino también comparar proporciones dentro de cada grupo. En este tipo de gráfico, cada barra representa una categoría principal (por ejemplo, grupos de tratamiento), y se divide en segmentos que corresponden a una segunda variable categórica, como la presencia o ausencia de diabetes.
En el caso de los datos clínicos, este gráfico resulta especialmente útil para analizar cómo se distribuye una condición dentro de diferentes grupos. Por ejemplo, al graficar diabetes por grupo, cada barra representa un grupo de pacientes, y dentro de ella podemos observar qué proporción tiene diabetes y cuál no. Esto permite identificar de forma visual si alguno de los grupos presenta mayor carga de enfermedad.
A diferencia del gráfico de barras simple, aquí utilizamos proporciones (por ejemplo, position = "fill" en ggplot), lo que normaliza todas las barras a la misma altura. De esta manera, podemos comparar directamente la distribución relativa de diabetes entre grupos, independientemente del número total de pacientes en cada uno.
Este tipo de gráfico sirve para:

Comparar proporciones de una variable categórica dentro de grupos
Detectar diferencias en prevalencia entre categorías
Facilitar la interpretación de relaciones entre variables cualitativas

Se recomienda utilizar barras apiladas cuando:

Se tienen dos variables categóricas
Se desea comparar proporciones y no solo frecuencias
Se busca identificar patrones de distribución entre grupos

```r
g_diabetes <- ggplot(datos, aes(x = grupo, fill = diabetes)) +
  geom_bar(position = "fill") +
  labs(
    title = "Proporción de diabetes por grupo",
    x = "Grupo",
    y = "Proporción",
    fill = "Diabetes"
  ) +
  theme_minimal()

g_diabetes

ggsave("figuras/02_diabetes_por_grupo.png", g_diabetes, width = 7, height = 5, dpi = 300)
```

```r
g_diabetes_colores <- ggplot(datos, aes(x = grupo, fill = diabetes)) +
  geom_bar(position = "fill") + # 'position = "fill"' para que muestre proporciones
  labs(
    title = "Proporción de diabetes por grupo (con colores personalizados)",
    x = "Grupo",
    y = "Proporción",
    fill = "Diabetes" # Etiqueta para la leyenda del color
  ) +
  theme_minimal() +
  # Usa scale_fill_manual para asignar colores a las categorías de diabetes.
  # Asegúrate de que 'No' y 'Sí' coincidan con los niveles de tu factor 'diabetes'.
  scale_fill_manual(values = c("No" = "lightgreen", "Sí" = "darkred")) # Asigna un color para 'No' y otro para 'Sí'

g_diabetes_colores # Muestra el gráfico

ggsave("figuras/04_diabetes_por_grupo_colores.png", g_diabetes_colores, width = 7, height = 5, dpi = 300) # Guarda el gráfico
```

#**Histograma**


El histograma es un tipo de gráfica utilizada para visualizar la distribución de una variable numérica. A diferencia del gráfico de barras, que se usa para categorías, el histograma agrupa los valores continuos en intervalos (llamados “bins”) y muestra cuántos datos caen dentro de cada uno.
En el caso de la glucosa, el histograma permite observar cómo se distribuyen los valores de los pacientes. Por ejemplo, nos ayuda a identificar si la mayoría de los pacientes tiene niveles normales, si existe una concentración en valores elevados, o si hay presencia de valores extremos que podrían ser clínicamente relevantes.
Este tipo de gráfico es especialmente útil porque nos permite:

Identificar la forma de la distribución (normal, sesgada, bimodal, etc.)
Detectar valores atípicos o extremos
Evaluar la variabilidad de los datos
Tener una idea general del comportamiento de la variable antes de aplicar análisis estadísticos

En el contexto clínico, analizar la distribución de la glucosa es fundamental, ya que permite detectar posibles patrones asociados a enfermedades como la diabetes o evaluar si los valores se encuentran dentro de rangos esperados.
Se recomienda usar histogramas cuando:

Se trabaja con variables numéricas continuas
Se desea explorar la distribución de los datos
Se está en una etapa inicial del análisis

```r
g_hist_glucosa <- ggplot(datos, aes(x = glucosa_mg_dl)) +
  geom_histogram(bins = 25, na.rm = TRUE) +
  labs(
    title = "Distribución de glucosa",
    x = "Glucosa (mg/dL)",
    y = "Frecuencia"
  ) +
  theme_minimal()

g_hist_glucosa

ggsave("figuras/03_histograma_glucosa.png", g_hist_glucosa, width = 7, height = 5, dpi = 300)
```

```r
g_hist_glucosa_color <- ggplot(datos, aes(x = glucosa_mg_dl)) +
  geom_histogram(bins = 25, na.rm = TRUE, fill = "skyblue", color = "black") + # Cambia 'skyblue' por el color de relleno que quieras y 'black' por el color del borde
  labs(
    title = "Distribución de glucosa (con color personalizado)",
    x = "Glucosa (mg/dL)",
    y = "Frecuencia"
  ) +
  theme_minimal()

g_hist_glucosa_color

ggsave("figuras/05_histograma_glucosa_color.png", g_hist_glucosa_color, width = 7, height = 5, dpi = 300)

# Explicación:
# - `fill = "skyblue"`: Establece el color de relleno de las barras del histograma. Puedes usar nombres de colores (como "red", "blue", "green") o códigos hexadecimales (como "#FF0000").
# - `color = "black"`: Establece el color del borde de las barras. Esto ayuda a definir visualmente cada "bin".
```

#**Boxplot: hemoglobina posterior por grupo**

El boxplot, o diagrama de cajas, es una gráfica utilizada para comparar la distribución de una variable numérica entre diferentes grupos. Este tipo de visualización resume la información en cinco elementos clave: el valor mínimo, el primer cuartil, la mediana, el tercer cuartil y el valor máximo, además de mostrar posibles valores atípicos.
En este caso, al graficar la hemoglobina posterior (hb_post) por grupo, podemos observar cómo se distribuyen los valores en cada grupo de tratamiento o control. Esto permite identificar rápidamente si existen diferencias entre grupos, como cambios en la mediana, mayor variabilidad o presencia de valores extremos.
El boxplot es especialmente útil en el contexto clínico porque:

Permite comparar resultados entre diferentes grupos (por ejemplo, tratamientos)
Muestra la variabilidad de los datos, no solo el promedio
Identifica valores atípicos que pueden ser clínicamente relevantes
Facilita la interpretación visual de posibles efectos del tratamiento

A diferencia de otras gráficas, el boxplot no muestra todos los datos individualmente, pero ofrece un resumen muy eficiente de su distribución, lo cual es ideal para comparaciones rápidas.
Se recomienda utilizar un boxplot cuando:

Se desea comparar una variable numérica entre dos o más grupos
Se quiere evaluar la dispersión y la mediana de los datos
Se busca detectar valores atípicos

Para cambiar los colores de los boxplots en `ggplot2`, puedes seguir un enfoque similar al de los gráficos de barras, utilizando el argumento `fill` dentro de `aes()` y `scale_fill_manual()` para asignar colores específicos a cada categoría de la variable que usas para agrupar (en este caso, `grupo`).

```r
g_box_hb <- ggplot(datos, aes(x = grupo, y = hb_post)) +
  geom_boxplot(na.rm = TRUE) +
  labs(
    title = "Hemoglobina posterior por grupo",
    x = "Grupo",
    y = "Hemoglobina posterior"
  ) +
  theme_minimal()

g_box_hb

ggsave("figuras/04_boxplot_hb_post_grupo.png", g_box_hb, width = 7, height = 5, dpi = 300)
```

```r
g_box_hb_colores <- ggplot(datos, aes(x = grupo, y = hb_post, fill = grupo)) + # Añadimos 'fill = grupo' para colorear los boxplots
  geom_boxplot(na.rm = TRUE) +
  labs(
    title = "Hemoglobina posterior por grupo (con colores)",
    x = "Grupo",
    y = "Hemoglobina posterior"
  ) +
  theme_minimal() +
  # Usa scale_fill_manual para asignar colores. Aquí usamos nombres de colores, similar a los gráficos de barras.
  scale_fill_manual(values = c("Control" = "steelblue", "Tratamiento_A" = "darkgreen", "Tratamiento_B" = "orange"))

g_box_hb_colores

ggsave("figuras/05_boxplot_hb_post_grupo_colores.png", g_box_hb_colores, width = 7, height = 5, dpi = 300)

# Explicación:
# - `fill = grupo`: Dentro de `aes()`, le indica a ggplot2 que use la variable `grupo` para rellenar (colorear) cada boxplot.
# - `scale_fill_manual(values = c(...))`: Permite especificar manualmente los colores para cada nivel del factor `grupo`. Asegúrate de que los nombres de los grupos ('Control', 'Tratamiento_A', 'Tratamiento_B') coincidan exactamente con los niveles de tu variable `grupo`.
```

#**Gráfico de dispersión: relación entre IMC y presión sistólica**
El gráfico de dispersión es una herramienta utilizada para analizar la relación entre dos variables numéricas. En este tipo de gráfica, cada punto representa una observación (por ejemplo, un paciente), ubicándose según sus valores en el eje X y el eje Y.
En este caso, al graficar el IMC (índice de masa corporal) en el eje X y la presión sistólica en el eje Y, podemos explorar si existe alguna relación entre estas dos variables. Por ejemplo, podemos observar si a mayor IMC tiende a aumentar la presión arterial, lo cual tiene una gran relevancia clínica.
Este tipo de gráfico es muy útil porque nos permite:

Detectar relaciones o asociaciones entre variables
Identificar tendencias (positivas, negativas o ausencia de relación)
Observar la dispersión de los datos
Detectar posibles valores atípicos

Además, al añadir una línea de tendencia (como una regresión lineal), podemos resumir la relación general entre las variables, lo que ayuda a interpretar si existe una asociación clara.
En el contexto clínico, este tipo de análisis es fundamental, ya que muchas investigaciones buscan precisamente identificar relaciones entre factores de riesgo y resultados en salud. Por ejemplo, explorar la relación entre IMC y presión sistólica puede ayudar a entender mejor el impacto del sobrepeso en la salud cardiovascular.
Se recomienda utilizar gráficos de dispersión cuando:

Se tienen dos variables numéricas
Se desea evaluar si existe una relación entre ellas
Se quiere explorar patrones o tendencias en los datos

## Explicación de `geom_point()` y `geom_smooth()`

En `ggplot2`, estas dos funciones son capas (geoms) que se añaden a un gráfico para representar diferentes aspectos de los datos.

*   ### `geom_point(na.rm = TRUE)`
    *   **¿Qué hace?** Esta función añade puntos individuales al gráfico, creando un gráfico de dispersión. Cada punto representa una observación en tus datos, con su posición determinada por los valores de las variables mapeadas a `x` e `y` en `aes()`.
    *   **`na.rm = TRUE`:** Este argumento significa "remove NA (Not Available)", es decir, "eliminar valores faltantes". Cuando `na.rm = TRUE`, `ggplot2` ignorará cualquier fila que contenga valores `NA` para las variables utilizadas en este `geom` (`imc` o `presion_sistolica` en este caso) al dibujar los puntos. Si no se incluye o se establece en `FALSE`, `ggplot2` mostrará una advertencia si encuentra `NA`s y no dibujará esos puntos.

*   ### `geom_smooth(method = "lm", se = TRUE, na.rm = TRUE)`
    *   **¿Qué hace?** Esta función añade una línea suavizada (una línea de tendencia) al gráfico, junto con su banda de confianza. Su propósito es visualizar la relación general entre las variables `x` e `y` sin ser demasiado sensible al ruido en los datos.
    *   **`method = "lm"`:** Especifica el método que se utilizará para calcular la línea suavizada. "lm" significa "linear model" (modelo lineal), lo que hace que `geom_smooth` ajuste una línea de regresión lineal simple a los datos.
    *   **`se = TRUE`:** Significa "standard error" (error estándar). Cuando es `TRUE` (que es el valor por defecto), `geom_smooth` dibujará una banda de confianza alrededor de la línea de tendencia. Esta banda te da una idea de la incertidumbre en la estimación de la línea. Un ancho menor de la banda sugiere una estimación más precisa.
    *   **`na.rm = TRUE`:** Al igual que en `geom_point`, este argumento indica que se deben ignorar los valores `NA` al calcular y dibujar la línea de tendencia y la banda de confianza.

```r
g_dispersion <- ggplot(datos, aes(x = imc, y = presion_sistolica)) +
  geom_point(na.rm = TRUE) +
  geom_smooth(method = "lm", se = TRUE, na.rm = TRUE) +
  labs(
    title = "Relación entre IMC y presión sistólica",
    x = "IMC",
    y = "Presión sistólica"
  ) +
  theme_minimal()

g_dispersion

ggsave("figuras/05_imc_presion.png", g_dispersion, width = 7, height = 5, dpi = 300)
```

## Personalizar el gráfico de dispersión: colores y formas de los puntos

Para cambiar el color o la forma de los puntos en un gráfico de dispersión, puedes usar los argumentos `color` o `shape` dentro de `aes()` y luego `scale_color_manual()` o `scale_shape_manual()` para asignar valores específicos. También puedes asignar el color o la forma en base a otra variable categórica de tus datos.

```r
# Gráfico de dispersión con puntos de color personalizado fijo (todos los puntos del mismo color)
g_dispersion_color_fijo <- ggplot(datos, aes(x = imc, y = presion_sistolica)) +
  geom_point(color = "darkblue", na.rm = TRUE) + # Color fijo para todos los puntos
  geom_smooth(method = "lm", se = TRUE, na.rm = TRUE) +
  labs(
    title = "Relación entre IMC y presión sistólica (color fijo)",
    x = "IMC",
    y = "Presión sistólica"
  ) +
  theme_minimal()

g_dispersion_color_fijo
```

```r
# Gráfico de dispersión con puntos coloreados según una variable (ej. sexo)
g_dispersion_color_variable <- ggplot(datos, aes(x = imc, y = presion_sistolica, color = sexo)) + # 'color = sexo' para colorear por la variable sexo
  geom_point(na.rm = TRUE) +
  geom_smooth(method = "lm", se = TRUE, na.rm = TRUE) +
  labs(
    title = "Relación entre IMC y presión sistólica (por sexo)",
    x = "IMC",
    y = "Presión sistólica",
    color = "Sexo" # Etiqueta para la leyenda de color
  ) +
  theme_minimal() +
  scale_color_manual(values = c("Femenino" = "purple", "Masculino" = "Blue")) # Asigna colores específicos a "F" y "M"

g_dispersion_color_variable
ggsave("figuras/07_imc_presion_color_sexo.png", g_dispersion_color_variable, width = 7, height = 5, dpi = 300)

# Explicación:
# - `color = sexo`: Dentro de `aes()`, le dice a ggplot2 que use la variable `sexo` para determinar el color de cada punto.
# - `scale_color_manual(values = c(...))`: Permite asignar colores específicos a cada nivel de la variable `sexo`.
```

```r
# Gráfico de dispersión con formas de puntos personalizadas según una variable (ej. diabetes)
g_dispersion_forma_variable <- ggplot(datos, aes(x = imc, y = presion_sistolica, shape = diabetes)) + # 'shape = diabetes' para cambiar la forma por la variable diabetes
  geom_point(size = 3, na.rm = TRUE) + # Aumentamos el tamaño de los puntos para que la forma sea más visible
  geom_smooth(method = "lm", se = TRUE, na.rm = TRUE) +
  labs(
    title = "Relación entre IMC y presión sistólica (por diabetes)",
    x = "IMC",
    y = "Presión sistólica",
    shape = "Diabetes" # Etiqueta para la leyenda de forma
  ) +
  theme_minimal() +
  scale_shape_manual(values = c("No" = 16, "Sí" = 17)) # Asigna formas específicas: 16 (círculo), 17 (triángulo)

g_dispersion_forma_variable


# Explicación:
# - `shape = diabetes`: Dentro de `aes()`, le dice a ggplot2 que use la variable `diabetes` para determinar la forma de cada punto.
# - `scale_shape_manual(values = c(...))`: Permite asignar formas específicas a cada nivel de la variable `diabetes`. Puedes encontrar una lista de códigos de formas en la documentación de ggplot2 o buscando "ggplot2 point shapes".
# - `size = 3`: Aumenta el tamaño de los puntos para que las formas sean más distinguibles.
```

```r
# Gráfico de dispersión combinando color y forma por variables
g_dispersion_color_forma_combinada <- ggplot(datos, aes(x = imc, y = presion_sistolica, color = sexo, shape = diabetes)) +
  geom_point(size = 3, na.rm = TRUE) +
  geom_smooth(method = "lm", se = TRUE, na.rm = TRUE, aes(color = sexo)) + # Color de la línea de suavizado también por sexo
  labs(
    title = "Relación IMC y presión sistólica (color por sexo, forma por diabetes)",
    x = "IMC",
    y = "Presión sistólica",
    color = "Sexo",
    shape = "Diabetes"
  ) +
  theme_minimal() +
  scale_color_manual(values = c("Femenino" = "purple", "Masculino" = "blue")) +
  scale_shape_manual(values = c("No" = 1, "Sí" = 4)) # Formas: 1 (círculo vacío), 4 (cruz)

g_dispersion_color_forma_combinada

# Explicación:
# - Aquí se combinan los mapeos `color = sexo` y `shape = diabetes` dentro de `aes()` para visualizar tres variables simultáneamente.
# - `geom_smooth(..., aes(color = sexo))`: Si quieres que la línea de tendencia también se adapte a cada grupo (en este caso, por `sexo`), debes incluir el `aes(color = sexo)` dentro de `geom_smooth`.
```

#**Facetas: distribución de glucosa por sexo y diabetes**

Las facetas son una herramienta avanzada de visualización que permiten dividir un gráfico en múltiples paneles según los niveles de una o más variables categóricas. Esto facilita el análisis simultáneo de subgrupos dentro de los datos, permitiendo comparaciones más detalladas y específicas.
En este caso, al graficar la glucosa utilizando facetas por sexo y estado de diabetes, estamos separando la información en distintos paneles que representan cada combinación de estas variables. Es decir, podemos observar cómo se distribuyen los niveles de glucosa en hombres y mujeres, tanto en pacientes con diabetes como sin ella.
Este tipo de visualización es especialmente útil en el ámbito clínico, ya que muchas veces no basta con analizar una variable de forma general. Las facetas permiten:

Comparar distribuciones entre múltiples grupos al mismo tiempo
Identificar patrones específicos dentro de subpoblaciones
Detectar diferencias que podrían no ser visibles en un gráfico general
Facilitar un análisis más detallado y estratificado

Por ejemplo, con este gráfico podemos evaluar si los pacientes con diabetes presentan mayores niveles de glucosa en ambos sexos, o si existe alguna diferencia entre hombres y mujeres dentro de cada grupo.
Se recomienda utilizar facetas cuando:

Se desea analizar una variable considerando varios factores al mismo tiempo
Se tienen subgrupos importantes para el análisis (como sexo, diagnóstico, tratamiento)
Se busca profundizar en la interpretación de los datos

```r
g_facetas <- ggplot(datos, aes(x = glucosa_mg_dl)) +
  geom_histogram(bins = 20, na.rm = TRUE) +
  facet_grid(diabetes ~ sexo) +
  labs(
    title = "Distribución de glucosa por diabetes y sexo",
    x = "Glucosa (mg/dL)",
    y = "Frecuencia"
  ) +
  theme_minimal()

g_facetas

ggsave("figuras/06_facetas_glucosa.png", g_facetas, width = 8, height = 6, dpi = 300)
```

## Personalizar colores en Gráficos de Facetas

Para cambiar el color de los histogramas dentro de las facetas, puedes usar el argumento `fill` dentro de `geom_histogram()`. Si quieres que el color varíe según una de las variables que usas para facetar (o cualquier otra variable categórica), debes incluirla en el `aes()` de `ggplot()` o `geom_histogram()` y luego usar `scale_fill_manual()`.


```r
# Gráfico de facetas con histogramas de un color fijo
g_facetas_color_fijo <- ggplot(datos, aes(x = glucosa_mg_dl)) +
  geom_histogram(bins = 20, na.rm = TRUE, fill = "skyblue", color = "black") + # Color fijo para todos los histogramas
  facet_grid(diabetes ~ sexo) +
  labs(
    title = "Distribución de glucosa por diabetes y sexo (color fijo)",
    x = "Glucosa (mg/dL)",
    y = "Frecuencia"
  ) +
  theme_minimal()

g_facetas_color_fijo
```

```r
# Gráfico de facetas con histogramas coloreados por la variable 'diabetes'
g_facetas_color <- ggplot(datos, aes(x = glucosa_mg_dl, fill = diabetes)) + # 'fill = diabetes' para colorear por la variable diabetes
  geom_histogram(bins = 20, na.rm = TRUE, color = "black") + # 'color = "black"' para añadir un borde a las barras
  facet_grid(diabetes ~ sexo) +
  labs(
    title = "Distribución de glucosa por diabetes y sexo (con color por diabetes)",
    x = "Glucosa (mg/dL)",
    y = "Frecuencia",
    fill = "Diabetes" # Etiqueta para la leyenda de color
  ) +
  theme_minimal() +
  scale_fill_manual(values = c("No" = "lightgreen", "Sí" = "darkred")) # Colores personalizados para 'No' y 'Sí' de diabetes

g_facetas_color

```
